```python
# CALCULANDO O DETERMINANTE DE UMA MATRIZ COM A BIBLIOTECA NUMPY
# np.linalg.det()
import numpy as np
A = np.array([[1,2],[4,4]])
A
np.linalg.det(A)
```

```python
# CALCULANDO O MATRIZ INVERSA DE UMA MATRIZ COM A BIBLIOTECA NUMPY
# np.linalg.det()
np.linalg.inv(A)
```

#### SOLVE PARA SISTEMAS LINEARES USANDO A BIBLIOTECA NUMPY

** np.linalg.solve(matriz_coef,vetor_resultado) **

```python
# criando uma matriz
import numpy as np
A = np.array([[1,1,1],[1,2,2],[2,1,3]])
# criando vetor resultado
R = np.array([[6,9,11]]).T
#A
R
res = np.linalg.solve(A,R)
print("O resultado do sistema de equações lineares é: ",res)
print(f'O resultado do sistema é x1:{res[0]}, e x2:{res[1]} e x3:{res[2]}.')
```

```python
# EXEMPLO 1
# Importando a biblioteca numpy
import numpy as np
# importando o pacote ou modulo pyplot da matplotilib
import matplotlib.pyplot as plt
# criando um objeto ou dominio da funçao
fx = np.arange(1,10)
y = 10 * fx + 12
# y = m*(x) + b  - função linear - plotar uma linha
# plotar a função linear chamar a método plot
plt.plot(fx,y)
# Colocar título no gráfico
plt.title("Curso Python2.0 - grafico")
# Colocar título nos eixos, x e y
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
# Comando para mostrar na tela a curva. tem IDE's que não é necessário
plt.show()
```

```python
# EXEMPLO 2 - Plotar a curva  y(x) = 10 * (x)**2 + 12
# Importando a biblioteca numpy
fx = np.arange(1,10)
yx = 10*(fx)**2 + 12

 # plotar a curva pelo método SCATTER
plt.scatter(fx,yx)
plt.title("Grafico 2")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

```python
# EXEMPLO 2 - Plotar a curva  y(x) = 10 * (x)**2 + 12
# Importando a biblioteca numpy
fx = np.arange(1,10)
yx = 10*(fx)**2 + 12
plt.plot(fx,yx)
plt.title("Grafico 2")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.grid()
plt.show()
```

```python
# EXEMPLO 3 - REFAZER EXEMPLO 2 - Utilizando a curva do tipo scatter + linha
fx = np.arange(1,10)
yx = 10*(fx)**2 + 12
plt.scatter(fx,yx)
plt.plot(fx,yx)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title(" Grafico scatter+linhas")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xa = np.linspace(0,100,200)
ya = np.exp(xa)
# criar o objeto subplot para chamar o a funçao plot
plt.figure()
plt.subplot(221)
plt.plot(xa,ya,'g--')
plt.title('Agua')
plt.ylabel('y1')
plt.subplot(222)
plt.plot(xa,ya,'r--')
plt.subplot(223)
plt.plot(xa,ya,'b--')
plt.subplot(224)
plt.plot(xa,ya,'k--')
plt.show()
```

```python
# integrate.quad(nome_func,a,b) -- calcula a integral de uma função
import numpy as np
from scipy.integrate import quad

# criar uma funçao - funçao lambda - y = lambda variavel: expressao_matematica
fx = lambda x: x**2

#chamar a funçao e fornecer o argumento da função
fx(10)


#Exemplo: resolvendo a integral x**2 dx no intervalo (0,10)
# quad(nome_funcao,limite_inf,limite_sup)
quad(fx,0,10)
```
