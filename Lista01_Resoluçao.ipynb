```python
""" Questão 13) Faça um programa que solicite ao usuário a sua idade e retorne a mensagem:
- idade < 12 - Você é uma criança;
- idade >= 12 e idade <= 20 - Você é um(a) adolecente;
- idade > 20 e idade <= 70 - Você é um(a) adulto(a);
- idade >= 70 - Você está na melhor idade"""

idade = int(input("Quantos anos você tem? "))
if(idade <= 12 and idade >= 0):
  print('Você é uma criança.')
elif(idade > 12 and idade <= 20):
  print('Você é um(a) adolecente.')
elif(idade >20 and idade <= 70):
    print('Você é um(a) adulto(a).')
elif(idade > 70):
  print('Você está na melhor idade.')
else:
  print('Número não especificado.\nTente novamente.')
```

```python
""" Questão 14) Refazer o exercício anterior utilize o laço de repetição while para executar o programa 4x.
- idade < 12 - Você é uma criança;
- idade >= 12 e idade <= 20 - Você é um(a) adolecente;
- idade > 20 e idade <= 70 - Você é um(a) adulto(a);
- idade >= 70 - Você está na melhor idade"""

iterador = 0
while(iterador < 4):
  idade = int(input("Quantos anos você tem? "))
  if(idade <= 12 and idade >= 0):
    print('Você é uma criança.')
  elif(idade > 12 and idade <= 20):
    print('Você é um(a) adolecente.')
  elif(idade >20 and idade <= 70):
      print('Você é um(a) adulto(a).')
  elif(idade > 70):
    print('Você está na melhor idade.')
  else:
    print('Número não especificado.\nTente novamente.')
  iterador +=1
```
