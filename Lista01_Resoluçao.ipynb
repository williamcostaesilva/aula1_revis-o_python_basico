```python
""" Quest√£o 1) Fa√ßa um algoritmo que calcule a √°rea de um c√≠rculo. Dica: pesquisar a biblioteca .math"""
# Area do c√≠rculo = pi * (raio)**2
# Importar a biblioteca matem√°tica
import math as mt
raio = 10 # cm
Area = mt.pi * (raio)**2
#Area
print("Area do circulo √©: ",Area)
print("Area do circulo √©: {}".format(Area))
print(f'Area do circulo √© {Area:.2f} cm.')
```

```python
""" Quest√£o 13) Fa√ßa um programa que solicite ao usu√°rio a sua idade e retorne a mensagem:
- idade < 12 - Voc√™ √© uma crian√ßa;
- idade >= 12 e idade <= 20 - Voc√™ √© um(a) adolecente;
- idade > 20 e idade <= 70 - Voc√™ √© um(a) adulto(a);
- idade >= 70 - Voc√™ est√° na melhor idade"""

idade = int(input("Quantos anos voc√™ tem? "))
if(idade <= 12 and idade >= 0):
  print('Voc√™ √© uma crian√ßa.')
elif(idade > 12 and idade <= 20):
  print('Voc√™ √© um(a) adolecente.')
elif(idade >20 and idade <= 70):
    print('Voc√™ √© um(a) adulto(a).')
elif(idade > 70):
  print('Voc√™ est√° na melhor idade.')
else:
  print('N√∫mero n√£o especificado.\nTente novamente.')
```

```python
""" Quest√£o 14) Refazer o exerc√≠cio anterior utilize o la√ßo de repeti√ß√£o while para executar o programa 4x.
- idade < 12 - Voc√™ √© uma crian√ßa;
- idade >= 12 e idade <= 20 - Voc√™ √© um(a) adolecente;
- idade > 20 e idade <= 70 - Voc√™ √© um(a) adulto(a);
- idade >= 70 - Voc√™ est√° na melhor idade"""

iterador = 0
while(iterador < 4):
  idade = int(input("Quantos anos voc√™ tem? "))
  if(idade <= 12 and idade >= 0):
    print('Voc√™ √© uma crian√ßa.')
  elif(idade > 12 and idade <= 20):
    print('Voc√™ √© um(a) adolecente.')
  elif(idade >20 and idade <= 70):
      print('Voc√™ √© um(a) adulto(a).')
  elif(idade > 70):
    print('Voc√™ est√° na melhor idade.')
  else:
    print('N√∫mero n√£o especificado.\nTente novamente.')
  iterador +=1
```
